```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
korea = pd.read_csv('C:\\GitHub\\test\\korea_2019.csv', encoding='cp949')
print(korea.head())
```

```python
korea_edit = korea.loc[:, ['가구별 일련번호', '가구원수', '맞벌이가구', '가구주_직업코드', '거처구분', '입주형태', '주택소유구분', '소득5분위코드(전국_전체가구)', '소득10분위코드(전국_전체가구)', '소득구간', '가계지출', '소비지출', '01. 식료품 · 비주류음료']]
korea_edit.head()
```

```python
korea_income = korea_edit.groupby(['소득구간']).agg({'가계지출': 'mean', '소비지출': 'mean', '01. 식료품 · 비주류음료': 'mean'}).reset_index()
print(korea_income.info())
```

```python
korea_income
```

```python
income_guess = korea_income['소득구간'] * 1000000
income_guess
```

```python
korea_income['소득추정'] = income_guess
korea_income
```

```python
engel = korea_income['01. 식료품 · 비주류음료'] / korea_income['소득추정'] * 100
korea_income['엥겔지수'] = engel
korea_income
```

```python
korea_income.to_csv('C:\\GitHub\\test\\korea_income.csv', encoding = 'utf-8-sig')
```

```python
korea_doubleincome = korea_edit.groupby(['소득구간', '맞벌이가구']).agg({'가계지출': 'mean', '소비지출': 'mean', '01. 식료품 · 비주류음료': 'mean'}).reset_index()
korea_doubleincome
```

```python
engel = korea_doubleincome['01. 식료품 · 비주류음료'] / (korea_doubleincome['소득구간'] * 1000000) *100 
korea_doubleincome['엥겔지수'] = engel
korea_doubleincome
```

```python
korea_doubleincome['맞벌이가구'] = korea_doubleincome['맞벌이가구'].astype('category')
korea_doubleincome['맞벌이가구'].cat.categories = ['외벌이', '맞벌이']
korea_doubleincome
```

```python
korea_doubleincome_compare = korea_doubleincome.groupby(['맞벌이가구']).엥겔지수.mean()
korea_doubleincome_compare
```
