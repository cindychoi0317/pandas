```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
korea = pd.read_csv('C:\\GitHub\\test\\korea_2019.csv', encoding='cp949')
print(korea.head())
```

```python
korea_edit = korea.loc[:, ['가구별 일련번호', '가구원수', '맞벌이가구', '가구주_직업코드', '거처구분', '입주형태', '주택소유구분', '소득5분위코드(전국_전체가구)', '소득10분위코드(전국_전체가구)', '소득구간', '가계지출', '소비지출', '01. 식료품 · 비주류음료']]
korea_edit.head()
```

```python
korea5 = korea_edit.groupby(['소득5분위코드(전국_전체가구)', '소득구간', '맞벌이가구']).agg({'가계지출': 'mean', '소비지출': 'mean', '01. 식료품 · 비주류음료': 'mean'}).reset_index()
print(korea5)
print(korea5.info())
print(korea5.index())
```

```python
income_guess = korea5['소득구간'] * 1000000
income_guess
```

```python
korea5['소득추정'] = income_guess
korea5
```

```python
engel = korea5['01. 식료품 · 비주류음료'] / korea5['소득추정'] * 100 
korea5['엥겔지수'] = engel
korea5
```

```python
korea5_income = korea5.groupby(['소득5분위코드(전국_전체가구)', '소득구간']).agg({'가계지출': 'mean', '소비지출': 'mean', '01. 식료품 · 비주류음료': 'mean'}).reset_index()
korea5_income['엥겔지수'] = engel
korea5_income
```

```python
korea5_doubleincome = korea5.groupby(['소득구간', '맞벌이가구']).agg({'가계지출': 'mean', '소비지출': 'mean', '01. 식료품 · 비주류음료': 'mean'}).reset_index()
korea5_doubleincome['엥겔지수'] = engel
korea5_doubleincome
```

```python
korea5_doubleincome['맞벌이가구'] = korea5_doubleincome['맞벌이가구'].astype("category")
```

```python
korea5_doubleincome['맞벌이가구'].cat.categories = ['외벌이', '맞벌이']
korea5_doubleincome
```

```python
korea5_doubleincome_compare = korea5_doubleincome.groupby(['맞벌이가구']).엥겔지수.mean()
korea5_doubleincome_compare
```
