```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a1 = pd.read_excel('C:\\GitHub\\test\\apt_Archive1\\all_2010_2013.xlsx')
a2 = pd.read_excel('C:\\GitHub\\test\\apt_Archive1\\all_2014_2016.xlsx')
a3 = pd.read_excel('C:\\GitHub\\test\\apt_Archive1\\all_2017_2020.xlsx')
```

```python
apt = pd.concat([a1, a2, a3])
apt.head()
```

```python
apt.info()
```

```python
sell = apt[(apt['category'] == '매매') & (apt['type'] == '아파트') & (apt['yy'] == 2020)]
sell
```

```python
ageage = 2020 - sell['year_built']
sell['age'] = ageage
sell.head()
```

```python
sell.info()
```

```python
sell['age_cut']=pd.cut(sell['age'],[0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 100], 
                      labels=['-5', '6-10', '11-15', '16-20', '21-25', '26-30', '31-35', '36-40', '41-45', '46-50', '51-55', '56-60', '61-'], duplicates='raise', include_lowest = True)
sell
```

```python
sell.groupby(['age_cut'])['age'].count()
```

```python
sell['q20'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.quantile(q=0.20, interpolation='nearest'))
sell['q40'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.quantile(q=0.40, interpolation='nearest'))
sell['q60'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.quantile(q=0.60, interpolation='nearest'))
sell['q80'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.quantile(q=0.80, interpolation='nearest'))
sell['med'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.median())
sell['mean'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.mean())
sell['min'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.min())
sell['max'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.max())
sell
```

```python
sell.loc[sell['price_sold'] <= sell['q20'], 'percentile'] = 'rank5'
sell.loc[(sell['price_sold'] > sell['q20']) & (sell['price_sold'] <= sell['q40']), 'percentile'] = 'rank4'
sell.loc[(sell['price_sold'] > sell['q40']) & (sell['price_sold'] <= sell['q60']), 'percentile'] = 'rank3'
sell.loc[(sell['price_sold'] > sell['q60']) & (sell['price_sold'] <= sell['q80']), 'percentile'] = 'rank2'
sell.loc[sell['price_sold'] > sell['q80'], 'percentile'] = 'rank1'
sell
```

```python
sell.sample(n=10)
```

```python
age_cut_percentile = sell.groupby(['age_cut', 'percentile'])['price_sold'].count().reset_index()
age_cut_percentile
```

```python
age_cut_percentile['price_sold'].sum()
```

```python
age_cut_percentile['price_sold'].astype(float)
```

```python
age_cut_percentile['sum'] = age_cut_percentile.groupby(['age_cut'])['price_sold'].transform(lambda x: x.sum())
age_cut_percentile
```

```python
age_cut_percentile['percent'] = age_cut_percentile['price_sold'] / age_cut_percentile['sum']
age_cut_percentile
```

```python
check = age_cut_percentile.groupby(['age_cut'])['percent'].sum()
check
```

```python
sell['min'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.min())
sell['p20'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.min() + (x.max() - x.min())/5*1)
sell['p40'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.min() + (x.max() - x.min())/5*2)
sell['p60'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.min() + (x.max() - x.min())/5*3)
sell['p80'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.min() + (x.max() - x.min())/5*4)
sell['max'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.max())
sell
```

```python
sell.loc[sell['price_sold'] <= sell['p20'], 'rank'] = 'rank5'
sell.loc[(sell['price_sold'] > sell['p20']) & (sell['price_sold'] <= sell['p40']), 'rank'] = 'rank4'
sell.loc[(sell['price_sold'] > sell['p40']) & (sell['price_sold'] <= sell['p60']), 'rank'] = 'rank3'
sell.loc[(sell['price_sold'] > sell['p60']) & (sell['price_sold'] <= sell['p80']), 'rank'] = 'rank2'
sell.loc[sell['price_sold'] > sell['p80'], 'rank'] = 'rank1'
sell
```

```python
age_cut_rank = sell.groupby(['age_cut', 'rank'])['price_sold'].count().reset_index()
age_cut_rank
```

```python
age_cut_rank['sum'] = age_cut_rank.groupby(['age_cut'])['price_sold'].transform(lambda x: x.sum())
age_cut_rank
```

```python
age_cut_rank['%'] = age_cut_rank['price_sold'] / age_cut_rank['sum'] * 100
age_cut_rank
```
