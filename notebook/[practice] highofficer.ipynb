```python
highofficer_rich_2019 = highofficer_rich[highofficer_rich['years']==2019]
highofficer_rich_2019
```

```python
q1 = highofficer_rich_2019['현재가액'].quantile(q = 0.2)
q2 = highofficer_rich_2019['현재가액'].quantile(q = 0.4)
q3 = highofficer_rich_2019['현재가액'].quantile(q = 0.6)
q4 = highofficer_rich_2019['현재가액'].quantile(q = 0.8)
q5 = highofficer_rich_2019['현재가액'].quantile(q = 1)
```

```python
print(q1)
print(q2)
print(q3)
print(q4)
print(q5)
```

```python
def q(x):
    if x > q4 and x <= q5:
        return '5분위'
    elif x > q3 and x <= q4:
        return '4분위'
    elif x > q2 and x <= q3:
        return '3분위'
    elif x > q1 and x <= q2:
        return '2분위'
    else:
        return '1분위'
```

```python
quantile = highofficer_rich_2019['현재가액'].apply(q)
highofficer_rich_2019['분위'] = quantile
highofficer_rich_2019
```
