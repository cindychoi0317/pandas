```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
apt = pd.read_excel('C:\\GitHub\\test\\apt_transaction.xlsx')
rent = pd.read_excel('C:\\GitHub\\test\\apt_rent_transaction.xlsx')
```

```python
apt.head()
```

```python
rent.head()
```

```python
apt_total = pd.concat([apt, rent], ignore_index = True)
apt_total
```

```python
type = apt_total['전월세구분'].fillna('매매')
deposit = apt_total['보증금(만원)'].fillna(0)
rent = apt_total['월세(만원)'].fillna(0)
pay = apt_total['거래금액(만원)'].fillna(0)
```

```python
apt_total['전월세구분'] = type
apt_total['보증금(만원)'] = deposit
apt_total['월세(만원)'] = rent
apt_total['거래금액(만원)'] = pay
apt_total
```

```python
apt_total.info()
```

```python
apt_total['구단위']=apt_total['시군구'].astype(str).str.split(' ').str[1]
apt_total['동단위']=apt_total['시군구'].astype(str).str.split(' ').str[2]
apt_total
```

```python
apt_total['거래금액(만원)'] = apt_total['거래금액(만원)'].str.strip()
apt_total['거래금액(만원)_re'] = apt_total['거래금액(만원)'].str.replace(',','')
apt_total
```

```python
apt_total.info()
```

```python
apt_total['거래금액(만원)_re'] = pd.to_numeric(apt_total['거래금액(만원)_re'])
apt_total['거래금액(만원)_re'].fillna(0, inplace = True)
apt_total
```

```python
apt_total.info()
```

```python
apt_total_buy = apt_total[apt_total['전월세구분'] == '매매']
apt_monthly = apt_total_buy.groupby(['구단위', '계약년월'])['거래금액(만원)_re'].mean().reset_index()
apt_monthly
```

```python
apt_monthly.to_excel('C:\\GitHub\\test\\apt_monthly.xlsx')
```

```python
apt_monthly_pivot = apt_monthly.pivot_table(
index = '구단위',
columns = '계약년월',
values = '거래금액(만원)_re',
dropna = False)

apt_monthly_pivot
```
