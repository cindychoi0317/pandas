```python
def ages(x):
    if x > 59:
        return '60대 이상'
    elif x > 49 and x < 60:
        return '50대'
    elif x > 39 and x < 50:
        return '40대'
    elif x > 29 and x < 40:
        return '30대'
    elif x > 19 and x < 30:
        return '20대'
    else:
        return '10대 이하'
```
