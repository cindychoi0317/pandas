```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
korea = pd.read_csv('C:\\GitHub\\test\\korea_2019.csv', encoding = 'cp949')
korea
```

```python
korea_edit = korea.loc[:, ['가구별 일련번호', '맞벌이가구', '소득구간', '거처구분', '입주형태', '전세보증금', '주거용 전용면적', '주택소유구분']]
korea_edit[['맞벌이가구', '거처구분', '입주형태', '주택소유구분']] = korea_edit[['맞벌이가구', '거처구분', '입주형태', '주택소유구분']].astype('category')
korea_edit['맞벌이가구'].cat.categories = ['외벌이', '맞벌이']
korea_edit['거처구분'].cat.categories = ['단독주택', '아파트', '3', '다세대', '5', '오피스텔', '7']
korea_edit['입주형태'].cat.categories = ['자가', '무상주택', '사택', '전세', '영구임대', '보증부월세', '7']
korea_edit['주택소유구분'].cat.categories = ['있음', '없음']
korea_edit
```

```python
korea_own_house = korea_edit.groupby(['소득구간', '주택소유구분']).agg({'가구별 일련번호': 'nunique', '전세보증금': 'mean', '주거용 전용면적': 'mean'}).reset_index()
korea_own_house
```

```python
portion = (korea_own_house['가구별 일련번호'] / korea_edit['가구별 일련번호'].max()) * 100
korea_own_house['가구수 비중'] = portion
korea_own_house = korea_own_house[['소득구간', '주택소유구분', '가구별 일련번호', '가구수 비중', '전세보증금', '주거용 전용면적']]
korea_own_house
```

```python
income_sum = korea_own_house.groupby('소득구간')['가구별 일련번호'].sum()
income_sum
```

```python
portion = korea_own_house['가구별 일련번호'] / income_sum
portion
```

```python
korea_own_house.to_csv('C:\\GitHub\\test\\korea_own_house.csv', encoding = 'utf-8-sig')
```
