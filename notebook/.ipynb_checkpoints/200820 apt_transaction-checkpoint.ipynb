```python
import pandas as pd
import numpy as np
```

```python
apt = pd.read_excel('C:\\GitHub\\test\\apt_Archive1\\apt_sell_all.xlsx')
apt.head()
```

```python
def size_re:
    if x < 66:
        return '20평 미만'
    elif x >= 66 | x < 99:
        return '20평'
    elif x >= 99 | x < 132:
        return '30평'
    else:
        return '40평 이상'
```

```python
apt['size_range']=pd.cut(apt['size'],[0, 65, 98, 132, 10000], labels=['20평 미만','20평대','30평대','40평 이상'], duplicates='raise')
apt.head(n=3)
```

```python
apt_yearly = apt.groupby(['town', 'yy'])['price_sold'].mean().reset_index()
apt_yearly
```

```python
apt_yearly_pivot = apt_yearly.pivot(index='town',columns='yy',values='price_sold')
apt_yearly_pivot
```

```python
apt_yearly_pivot.to_excel('C:\\GitHub\\test\\apt_Archive1\\apt_yearly_pivot.xlsx')
```

```python
apt_size_yearly = apt.groupby(['town', 'yy', 'size_range'])['price_sold'].mean().reset_index()
apt_size_yearly
```

```python
apt_size_yearly_pivot = pd.pivot_table(apt_size_yearly, index=['size_range','town'],columns='yy',values='price_sold')
apt_size_yearly_pivot
```

```python
apt_size_yearly_pivot.to_excel('C:\\GitHub\\test\\apt_Archive1\\apt_size_yearly_pivot.xlsx')
```
