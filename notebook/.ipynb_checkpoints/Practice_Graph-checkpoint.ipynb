```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gapminder = pd.read_csv('C:\\GitHub\\pandas\\data\\gapminder.tsv', sep= '\t')
print(gapminder)
print(gapminder['country'].unique())
```

```python
print(gapminder['continent'].nunique())
gapminder_continental = gapminder.groupby(['continent', 'year'])['gdpPercap'].mean()
print(gapminder_continental)
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
axes1.plot(gapminder['year'], gapminder['gdpPercap'])
axes1.set_xlabel('year')
axes1.set_ylabel('gdpPercap')
axes1.set_title('Yearly gdpPercap by Continent')
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
axes1.scatter(
    x = gapminder['year'], 
    y = gapminder['gdpPercap'])
axes1.set_title('Year and GdpPerCap')
axes1.set_xlabel('Year')
axes1.set_ylabel('GdpPerCap')
```

```python
gdpPercap_mean = gapminder.groupby('year', as_index = False)['gdpPercap', 'lifeExp'].mean()
print(gdpPercap_mean)

print(gdpPercap_mean.columns)
print(type(gdpPercap_mean))
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
axes1.scatter(
    x = gdpPercap_mean['year'], 
    y = gdpPercap_mean['gdpPercap'])
axes1.set_title('Year and GdpPerCap')
axes1.set_xlabel('Year')
axes1.set_ylabel('GdpPerCap')
```

```python
gdpPercap_yearly_continental = gapminder.groupby(['continent', 'year'], as_index = False)['gdpPercap'].mean()

def recode_continent(continent):
    if continent == 'Africa':
        return 0
    elif continent == 'Americas':
        return 1
    elif continent == 'Asia':
        return 2
    elif continent == 'Europe':
        return 3
    else:
        return 4 

gdpPercap_yearly_continental['continent_color'] = gdpPercap_yearly_continental['continent'].apply(recode_continent)
print(gdpPercap_yearly_continental)
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
axes1.scatter(
    x = gdpPercap_yearly_continental['year'],
    y = gdpPercap_yearly_continental['gdpPercap'],
    c = gdpPercap_yearly_continental['continent_color'])
```

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)

axes1.boxplot([gapminder[gapminder['continent'] == 'Africa']['pop'],
               gapminder[gapminder['continent'] == 'Americas']['pop'],
               gapminder[gapminder['continent'] == 'Asia']['pop'],
               gapminder[gapminder['continent'] == 'Europe']['pop'],
               gapminder[gapminder['continent'] == 'Oceania']['pop']],
               labels = ['Africa', 'Americas', 'Asia', 'Europe', 'Oceania'])
axes1.set_title('Pop by Continent')
axes1.set_xlabel('Continent')
axes1.set_ylabel('Pop')

fig
```
