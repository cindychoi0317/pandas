```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
```

```python
a1 = pd.read_excel('C:\\GitHub\\test\\apt_Archive1\\all_2010_2013.xlsx')
a2 = pd.read_excel('C:\\GitHub\\test\\apt_Archive1\\all_2014_2016.xlsx')
a3 = pd.read_excel('C:\\GitHub\\test\\apt_Archive1\\all_2017_2020.xlsx')
```

```python
apt = pd.concat([a1, a2, a3])
apt.head()
```

```python
apt.info()
```

```python
sell = apt[(apt['category'] == '매매') & (apt['type'] == '아파트')]
sell
```

```python
ageage = 2020 - sell['year_built']
sell['age'] = ageage
sell.head()
```

```python
sell.info()
```

```python
sell['age_cut']=pd.cut(sell['age'],[0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 100], 
                      labels=['-5', '6-10', '11-15', '16-20', '21-25', '26-30', '31-35', '36-40', '41-45', '46-50', '51-55', '56-60', '61-'], duplicates='raise', include_lowest = True)
sell
```

```python
apt[apt['age'] == 31]
```

```python
sell.groupby(['age_cut'])['age'].count()
```

```python
sell['q25'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.quantile(q=0.25, interpolation='nearest'))
sell['q50'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.quantile(q=0.50, interpolation='nearest'))
sell['q75'] = sell.groupby(['age_cut'])['price_sold'].transform(lambda x: x.quantile(q=0.75, interpolation='nearest'))
sell
```

```python
q25 = sell.groupby(['age_cut'])['price_sold'].quantile(q=0.25)
q50 = sell.groupby(['age_cut'])['price_sold'].quantile(q=0.50)
q75 = sell.groupby(['age_cut'])['price_sold'].quantile(q=0.75)
print(q25)
print(q50)
print(q75)
```
