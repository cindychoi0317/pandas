```python
p2p_edit = p2p.loc[:, ['경매번호', '자금모집종료일', '대출자ID', '투자모집 성공여부']]
```

```python
p2p_edit.head(n=3)
```

```python
p2p_edit['년'] = p2p_edit['자금모집종료일'].dt.year
p2p_edit['월'] = p2p_edit['자금모집종료일'].dt.month
```

```python
p2p_edit
```

```python
p2p_edit['투자모집 성공'] = np.where(p2p_edit['투자모집 성공여부'] == 1, '성공', '')
p2p_edit
```

```python
p2p_monthly = p2p_edit.groupby(['년', '월'])['경매번호'].count().reset_index()
p2p_monthly.rename(columns = {'경매번호' : '투자신청'}, inplace = True)
p2p_monthly
```

```python
success = p2p_edit.groupby(['년', '월'])['투자모집 성공'].count().fillna(0).reset_index()
p2p_monthly['투자성공'] = success
p2p_monthly
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ax = plt.subplots()
ax = sns.distplot(p2p_monthly['월'], hist = False)
ax.set_title()
```
