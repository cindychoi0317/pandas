```python
import pandas as pd
import numpy as np
```

```python
apt1 = pd.read_excel('C:\\GitHub\\test\\apt_Archive1\\all_2010_2013.xlsx')
apt2 = pd.read_excel('C:\\GitHub\\test\\apt_Archive1\\all_2014_2016.xlsx')
apt3 = pd.read_excel('C:\\GitHub\\test\\apt_Archive1\\all_2017_2020.xlsx')
```

```python
apt = pd.concat([apt1, apt2, apt3])
apt.head()
```

```python
apt.info()
```

```python
apt['age'] = 2020 - apt['year_built']
apt.head()
```

```python
apt_sell = apt[apt['category'] == '매매']
apt_own = apt_sell.groupby(['town', 'village', 'yy', 'mm', 'name', 'size', 'age'])['price_sold'].mean().reset_index()
apt_own
```

```python
rent1 = pd.read_excel('C:\\GitHub\\test\\apt_rent\\apt_rent_all_new_1.xlsx')
rent2 = pd.read_excel('C:\\GitHub\\test\\apt_rent\\apt_rent_all_new_2.xlsx')
```

```python
rent_all = pd.concat([rent1, rent2])
rent_all
```

```python
rent = rent_all[(rent_all['monthlyrent'] == 0) & (rent_all['category'] == '전세')]
rent
```

```python
apt_rent = rent.groupby(['town', 'village', 'yy', 'mm', 'name', 'size', 'age'])['deposit'].mean().reset_index()
apt_rent
```

```python
merge =  pd.merge(apt_own, apt_rent, how = 'left',  left_on = ['town', 'village', 'yy','mm','name','size','age'], right_on = ['town', 'village', 'yy','mm','name','size','age'])
merge
```

```python
merge[merge['price_sold'] == 0]
```

```python
merge['deposit'].fillna(0,inplace=True)
```

```python
merge
```

```python
gap = merge['price_sold'] - merge['deposit']
merge['gap'] = gap
merge
```

```python
gap_under2 = merge[(merge['gap'] <= 20000) & (merge['gap'] > 0) & (merge['deposit'] > 0)]
gap_under2
```
